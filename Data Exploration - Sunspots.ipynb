```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/sunspots.csv', sep=';', header=None)
df.columns = ['Year', 'Month', 'YearFrac', 'Mean', 'StDev', 'Observations', 'Definitive']
df
```

```python
# Clear null placeholders
df.loc[df.Mean < 0, 'Mean'] = np.nan
df.loc[df.StDev < 0, 'StDev'] = np.nan
df.loc[df.Observations < 0, 'Observations'] = np.nan
df.describe()
```

```python
import matplotlib.pyplot as plt

df.plot(x='YearFrac', y='Mean')
plt.show()
```
