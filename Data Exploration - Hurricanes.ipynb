```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/combined.csv', parse_dates=[2])
df
```

```python
df.dtypes
```

```python
print('Ends in 00:', (df.Time % 100 == 0).sum())
print('Ends in 15:', (df.Time % 100 == 15).sum())
print('Ends in 30:', (df.Time % 100 == 30).sum())
print('Ends in 45:', (df.Time % 100 == 45).sum())
(df.Time % 100).value_counts()
```

```python
df.loc[df['Maximum Wind'] < 0, 'Maximum Wind'] = np.nan
df.loc[df['Minimum Pressure'] < 0, 'Minimum Pressure'] = np.nan
for low_wind in ['Low Wind NE', 'Low Wind SE', 'Low Wind SW', 'Low Wind NW']:
    df.loc[df[low_wind] < 0, low_wind] = np.nan
for moderate_wind in ['Moderate Wind NE', 'Moderate Wind SE', 'Moderate Wind SW', 'Moderate Wind NW']:
    df.loc[df[moderate_wind] < 0, moderate_wind] = np.nan
for high_wind in ['High Wind NE', 'High Wind SE', 'High Wind SW', 'High Wind NW']:
    df.loc[df[high_wind] < 0, high_wind] = np.nan
df.describe()
```

```python
low_wind = ['Low Wind NE', 'Low Wind SE', 'Low Wind SW', 'Low Wind NW']
for lw in low_wind:
    df.loc[df[lw] < 0, lw] = np.nan
df.describe()
```
